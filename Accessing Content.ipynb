```python
def pretty_result_item(item):
    result = '\n'
    result += 'Title:\t{0}\n'.format( item.title )
    result += 'ID:\t{0}\n'.format( item.id )
    result += 'Type:\t{0}\n'.format( item.type )
    result += 'Owner:\t{0}\n'.format( item.owner )
    result += 'Tags:\t{0}\n'.format( ', '.join(item.tags) )
    result += 'Description:\t{0}\n'.format( item.description )
    print ( result )
```

```python
for item in search_result:
    pretty_result_item(item)
```

```python
search_result_google = gis.content.search(query="title:Google*")
for item in search_result_google:
    pretty_result_item(item)
```
